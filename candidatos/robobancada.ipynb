```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import unidecode
```

Função para retirar acentuação

```python
def f(str):
    return (unidecode.unidecode(str))
```

Raspa tabela de site

```python
link = 'http://fpagropecuaria.org.br/integrantes/todos-os-integrantes/'
url = urlopen(link)
bsObje = BeautifulSoup(url, "html5lib")
tabela = bsObje.findAll("tbody", {"class":"coresAlternadas"})[0]
```

Cria lista com os nomes

```python
nomes = []
```

```python
for table_ele in tabela.findAll('tr'):
    linha = table_ele.findAll('td')[0].text.strip()
    dicionario = {"nome_bancada_ruralista": linha}
    nomes.append(dicionario)     
```

Cria dataframe

```python
df_nomes = pd.DataFrame(nomes)
```

```python
df_nomes.drop(df_nomes.tail(1).index,inplace=True) # apaga última linha que tem Total
```

```python
df_nomes.info()
```

Limpeza

```python
df_nomes["nome_bancada_ruralista"] = df_nomes["nome_bancada_ruralista"].apply(f)
```

```python
def limpeza1(linha):
    nome = linha["nome_bancada_ruralista"]
    nome = nome.replace('*', '')
    return nome
```

```python
df_nomes["nome_bancada_ruralista"] = df_nomes.apply(limpeza1, axis=1)
```
