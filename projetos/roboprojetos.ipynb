```python
import pandas as pd
import requests
```

```python
# Documentação: dadosabertos.camara.leg.br/swagger/api.html
```

```python
prefixo = 'https://dadosabertos.camara.leg.br/api/v2/proposicoes/'
```

```python
#lista_projetos = pd.read_csv('dados/projetos_teste.csv', sep=',',encoding = 'utf-8', converters={'chave': lambda x: str(x)})

lista_projetos = pd.read_excel('dados/projetos_dia25julho.xlsx',sheet_name='PLs propostos')
lista_projetos.reset_index().head()
```

```python
lista_projetos.info()
```

```python
lista_projetos["chave"] = lista_projetos.link_projeto.str.split("=", 1).str[1]
```

```python
lista_projetos.reset_index().head()
```

```python
lista_projetos.info()
```

```python
projetos = []

for num, row in lista_projetos.iterrows():
    projeto = (row['n_projeto']).strip()
    print(projeto)
    chave = (row['chave']).strip()
    url = prefixo + chave + '/autores'
    
    parametros = {'formato': 'json', 'itens': 100}
    resposta = requests.get(url, parametros)
    for vez in resposta.json()['dados']:
        dicionario = {"parlamentar": vez['nome'], 
                          "uri": vez['uri'], 
                          "projeto": projeto, 
                          "tipo": vez['tipo'],
                          "cod_tipo": vez['codTipo']
                          }
        projetos.append(dicionario)
```

```python
df_projetos_api = pd.DataFrame(projetos)
```

```python
df_projetos_api.reset_index()
```

```python
#nome = df_projetos_api[(df_projetos_api['projeto'] == 'PL 6136/2016')]
#nome
```

```python
projetos_autores = []

for num, row in df_projetos_api.iterrows():
    projeto = (row['projeto']).strip()
    url = (row['uri'])
    autor = (row['parlamentar']).strip()
    #print(projeto + " de " + autor)
    tipo = (row['tipo']).strip()
    if tipo == 'Deputado' and url != None: # None são deputados de legislaturas passadas, que não têm uri
        parametros = {'formato': 'json', 'itens': 100}
        resposta = requests.get(url, parametros)
        dicionario = {"deputado_id": str(resposta.json()['dados']['id']),
                      "nome": autor, 
                      "uri": url, 
                      "projeto": projeto, 
                      "legislatura": resposta.json()['dados']['ultimoStatus']['idLegislatura'],
                      "situacao": resposta.json()['dados']['ultimoStatus']['situacao'],
                      "condicao": resposta.json()['dados']['ultimoStatus']['condicaoEleitoral']}
        projetos_autores.append(dicionario)
    else:
        dicionario = {"deputado_id": "não_deputado",
                          "nome": autor, 
                          "uri": tipo, 
                          "projeto": projeto, 
                          "legislatura": 'nao_deputado',
                          "situacao": 'nao_deputado'
                          }
        projetos_autores.append(dicionario)
```

```python
df_projetos_autores_api = pd.DataFrame(projetos_autores)
```

```python
df_projetos_autores_api.reset_index().head()
```

```python
df_projetos_autores_api.info()
```

```python
selecionados = df_projetos_autores_api[(df_projetos_autores_api['legislatura'] == 55) &
                                       (df_projetos_autores_api['condicao'] == 'Titular')]
```

```python
selecionados.reset_index()
```

```python
#nome = selecionados[(selecionados['projeto'] == 'PL 6136/2016')]
#nome
```

```python
so_projetos = selecionados['projeto'].value_counts()
so_projetos = pd.DataFrame({'projeto':so_projetos.index, 'quantidade':so_projetos.values})
```

```python
so_projetos.reset_index()
```

```python
def limpeza1(linha):
    documento = str(linha["n_projeto"])
    documento = documento.strip()
    return str(documento)
```

```python
lista_projetos["n_projeto"] = lista_projetos.apply(limpeza1, axis=1)
```

```python
df_projetos_eleitos = pd.merge(lista_projetos, so_projetos, left_on='n_projeto', right_on='projeto')
```

```python
df_projetos_eleitos.info()
```

```python
df_projetos_eleitos.drop('chave',axis=1,inplace=True)
df_projetos_eleitos.drop('projeto',axis=1,inplace=True)
df_projetos_eleitos.drop('quantidade',axis=1,inplace=True)
```

```python
df_projetos_eleitos.to_excel('resultados/projetos_dep_eleitos.xlsx',sheet_name='Sheet1')
```
