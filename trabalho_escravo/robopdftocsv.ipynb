```python
# Acessa lista de abril de 2018, já padronizada manualmente, para limpeza final
```

```python
df2 = pd.read_excel('pdf_to_csv/listate_abril2018_padronizada.xlsx',sheet_name='Planilha1')
```

```python
df2.reset_index().head()
```

```python
# Tira símbolos do documento
def limpeza1(linha):
    documento = str(linha["CNPJ_CPF_CEI"])
    documento = documento.strip()
    documento = documento.replace('.', '')
    documento = documento.replace('-', '')
    documento = documento.replace('/', '')
    return str(documento)  

# retira espaços em branco
def limpeza2(linha):
    documento = str(linha["cidade"])
    documento = documento.strip()
    return str(documento) 
```

```python
# Separa cidade e Estado
```

```python
df2[["cidade", "estado"]] = df2["Nome_propriedade"].apply(lambda x :x.split(",")[-1]).str.split("/", expand=True)
```

```python
# Apaga informações sobressalentes de Nome_propriedade - opcional
```

```python
df2["Nome_propriedade"] = df2["Nome_propriedade"].apply(lambda x :",".join(x.split(",")[:-1]))
```

```python
# Chama funações acima de limpeza
```

```python
df2["cidade"] = df2.apply(limpeza2, axis=1)
```

```python
df2["CPF_CNPJ_limpo"] = df2.apply(limpeza1, axis=1)
```

```python
df2.reset_index()
```

```python
df2.to_excel('pdf_to_csv/listate_abril2018_limpa.xlsx',sheet_name='Sheet1')
df2.to_csv('pdf_to_csv/listate_abril2018_limpa.csv', index=False)
```
