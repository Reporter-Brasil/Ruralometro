```python
import pandas as pd
import requests
```

```python
# Documentação: dadosabertos.camara.leg.br/swagger/api.html
```

```python
prefixo = 'https://dadosabertos.camara.leg.br/api/v2/votacoes/'
```

```python
lista_projetos = pd.read_excel('lista_projetos/exemplo.xlsx',sheet_name='Planilha1', converters={'chave': lambda x: str(x), 'data_votacao': lambda x: str(x)})
```

```python
votos = []

for num, row in lista_projetos.iterrows():
    projeto = (row['projeto']).strip()
    print(projeto)
    data_votacao = (row['data_votacao']).strip()
    avaliado_por = (row['avaliado_por']).strip()
    avaliacao = (row['avaliacao']).strip()
    relevancia = (row['relevancia']).strip()
    chave = (row['chave']).strip()
    url = prefixo + chave + '/votos'
    for pagina in [1, 2, 3, 4, 5, 6, 7]:
        parametros = {'formato': 'json', 'itens': 100, 'pagina': pagina}
        resposta = requests.get(url, parametros)
        for voto in resposta.json()['dados']:
            #print(voto)
            dicionario = {"deputado": voto['parlamentar']['nome'], 
                          "como_votou": voto['voto'], 
                          "partido_votacao": voto['parlamentar']['siglaPartido'], 
                          "uf": voto['parlamentar']['siglaUf'], 
                          "deputado_id": str(voto['parlamentar']['id']),
                          "projeto": projeto, 
                          "data_votacao": data_votacao, 
                          "avaliado_por": avaliado_por, 
                          "avaliacao": avaliacao, 
                          "relevancia": relevancia}
            votos.append(dicionario)
```

```python
df_votos_api = pd.DataFrame(votos)
```

```python
df_votos_api.reset_index()
```
